```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import PIL 
import skimage.io as io
```

```python
os.getcwd()
```

```python
# load pngs:
clevr_comps = io.imread('videos/clevr_comet_reconstructed_best.png')
clevr_gifs = io.imread('videos/clevr_comet_best.gif')
our_clevr_comps = io.imread('videos/Clevr_reconstructed_our_code.png')
our_clevr_gifs = io.imread('videos/Clevr_our_code.gif')
```

```python
clevr_gifs = clevr_gifs[-1]
our_clevr_gifs = our_clevr_gifs[-1]
print(clevr_gifs.shape)
print(our_clevr_gifs.shape)
```

```python
# plot a 2x2 grid with clevr at the left and our clevr at the right
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].imshow(clevr_comps)
axs[0, 0].set_title('COMET components')
axs[0, 1].imshow(our_clevr_comps)
axs[0, 1].set_title('Refactored Code components')
axs[1, 0].imshow(clevr_gifs)
axs[1, 0].set_title('COMET reconstruction')
axs[1, 1].imshow(our_clevr_gifs)
axs[1, 1].set_title('Refactored Code reconstruction')
axs[0, 0].axis('off')
axs[0, 1].axis('off')
axs[1, 0].axis('off')
axs[1, 1].axis('off')
plt.tight_layout()
plt.savefig('videos/compare_clevr_comet.png', dpi=300)
plt.show()
# load pngs:
```
